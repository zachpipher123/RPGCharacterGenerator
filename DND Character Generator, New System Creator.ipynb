```python
import random
```

```python
class character(object):
    def __init__(self, strength, magic, tech):
        health = 100
        self.strength = strength
        self.magic = magic
        self.tech = tech
```

```python
character_sheet = []
for x in range(1000):
    character_sheet.append(character(strength = random.randint(1,10), magic = random.randint(1, 10), tech = random.randint(1,10)))
    if character_sheet[-1].strength == 10 and character_sheet[-1].magic == 10 and character_sheet[-1].tech == 10:
        print("Perfect Character")
```

```python
#This is a rough draft how to calculate how fast a character would kill a target
#This general idea is the frame work how a character will choose spells and abilities or just to auto attack
#Then you can sort characters by how fast they killed the target 
time_kill = []
counter = 0
target_dummy = 100
for index, x in enumerate(character_sheet):
    while target_dummy > 0:
        target_dummy = target_dummy - x.strength
        counter = counter + 1
    time_kill.append((counter, x.strength))
    target_dummy = 100
    counter = 0
    print(time_kill[index])
```

```python
time_kill.sort()
```
